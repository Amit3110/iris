```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

---

iris dataset is already available in SciKit learn library and we can directly import it

```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
iris_data=iris.data
iris_data=pd.DataFrame(iris_data,columns=iris.feature_names)
iris_data['class']=iris.target
iris_data.head()
```

Applying the algorithm

```python
#Train test split
from sklearn.model_selection import train_test_split
X=iris_data.values[:,0:4]
Y=iris_data.values[:,4]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

Training the model

Algo used:- KNN, Support Vector Machine(SVM), Randomforest, Logistic Regression

```python
#KNN
model=KNeighborsClassifier()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(accuracy_score(y_test,predictions))

#It gives around 0.9466666666667
```

```python
#SVM
model=SVC()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(accuracy_score(y_test,predictions))

#It gives around 0.98666666666667
```

```python
#Randomforest
model=RandomForestClassifier(n_estimators=5)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(accuracy_score(y_test,predictions))

#It gives around 0.975903614478314
```

```python
#LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(accuracy_score(y_test,predictions))

#It gives around 0.963855421686747
```
